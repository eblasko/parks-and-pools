```python
import pandas as pd
import numpy as np
import geopandas as gpd
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
parks = pd.read_csv("playgrounds.csv")
u = parks.groupby('neighborhood').count().loc[:,"id"].sort_values(ascending = False)
u
```

```python
u[0:10].plot.bar()
```

```python
pep = pd.read_csv("pequipment.csv")
pepnum = pep.groupby('neighborhood').count().loc[:,"equipment_type"].sort_values(ascending = False)
pepnum
```

```python
pep_series = pep.groupby('neighborhood').count().loc[:,"equipment_type"].sort_values()
pep_series[56:66].plot.bar()
```

```python
pavg = pepnum/u
pavg.sort_values(ascending = False)
```

```python
ppavg = pavg.sort_values(ascending = False)
ppavg[0:4].plot.pie()
```

```python
pscore = (u*(1+(pavg/10)))
pscore.sort_values(ascending = False)
```

```python
ppscore = pscore.sort_values()
ppscore[0:66].plot.bar(figsize=(20,10))
```
