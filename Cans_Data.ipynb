```python
import pandas as pd



def filter_less_than(x, threshold):
    if len(x) > threshold:
        return True
    else: 
        return False

data = pd.read_csv("cans.csv")
data = data.groupby("neighborhood").filter(filter_less_than, threshold = 20)

data.groupby("neighborhood").count().sort_values("container_id")
```
